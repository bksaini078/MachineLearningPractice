```python
import csv
import numpy as np
import pandas as pd

# We use sklearn for preprocessing. Do not use it in your code!!!
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler

# Below is an example how data preprocessing can be achieved with pandas module
data = pd.DataFrame()
test = pd.DataFrame()

data = pd.read_csv('trainData.csv',  dtype={'Color': str,'Radius (cm)':float,'Weight (grams)':float})
# print(data)
#########################  DATA Preprocessing  #############################
    
# drop duplicated rows
data= data.drop_duplicates()
    
# replace zeros entireis with np.nan
data=data.replace(0,np.nan)
data=data.dropna() # drop all nan entiries 
    
# transform the nominal feature (color) to dummy
newCols=pd.get_dummies(data.iloc[:,0]) 
data = pd.concat([newCols, data], axis=1) # add them to the transfomed columns to the beggining  of the data frame
data=data.drop(data.columns[3],axis = 1) # drop nominal column (color)    

## normalization
min_radius = min(data.iloc[:,3])
max_radius = max(data.iloc[:,3])
data['Radius (cm)'] = list(map(lambda x: (float(x)-min_radius)/(max_radius-min_radius), data.iloc[:,3]))

## normalization
min_weight = min(data.iloc[:,4])
max_weight = max(data.iloc[:,4])
data['Weight (grams)'] = list(map(lambda x: (float(x)-min_weight)/(max_weight-min_weight), data.iloc[:,4]))
train=data.copy()
#reset indexing as some index is missing index 8
train=train.reset_index(drop=True)
train
```

```python
# TODO
#function to calculate euclidean distance 
def Euclidean_Distance(test_set,train_set,x,column_list):
    train_set['Euclidean_Distance']=0
    for i in column_list:
        train_set['Euclidean_Distance']+= np.square(train_set.loc[:,i]-test_set.loc[x,i])
    train_set['Euclidean_Distance']=np.sqrt(train_set['Euclidean_Distance'])
    return train_set

#function for K nearest neighbours
def KNN_Algorithm(test, train1,k,column_list):
    temp= train1.copy()
    test1=test.copy()
  
    for i in range(0,len(test1)):
        #calculating euclidean distance
        temp=Euclidean_Distance(test1,train1,i,column_list)
    
#         taking k smallest values as nearest neigbhours
        N_Neighbour=temp.nsmallest(k,'Euclidean_Distance')
       
        #sorting according to size for dealing with K values 
        N_Neighbour=N_Neighbour.groupby('Fruit (class)').size().reset_index(name='size').sort_values('size',ascending= False)
        N_Neighbour=N_Neighbour.reset_index()
        
#         assigning predicted values 
        test1.loc[i,'Predicted (class)']=N_Neighbour.loc[0,'Fruit (class)']
       
    return test1

#performance calculation: precision 
def Performance_Evaluation(data):
    true_Positive = data.where(data['Predicted (class)']==data['Original (class)']).count().unique()[0]
    false_Positive= data.where(data['Predicted (class)']!=data['Original (class)']).count().unique()[0]
    precision=true_Positive/(false_Positive+true_Positive)*100
    return precision

```

```python
#taking input 
k= int(input("Please enter value of K for KNN algorithm:"))

#list of column for prediction , for euclidean distance, as they are vectors
column_list = ["Green","Red","Yellow","Radius (cm)","Weight (grams)"]


predicted_result_k = KNN_Algorithm(test,train,k,column_list)
precision_k = Performance_Evaluation(predicted_result_k)



print("Precision @",k,":", precision_k)



predicted_result_k
```
