# Logistic Regression

__Difference between Regression and Classification.__<br>
Regression- Quantitative 
Classification- Qualitative (eye color, yes/no, spam/not spam)

__Why can't we solve qualitative problem with regression?__ <br>
lets take an example <br>
x1 x2 y <br>
1  2  0 <br>
2  1  1 <br>
3  4  2 <br>
6  7  0 <br>

In above dataset, the y has three classes 0,1,2. If we train the linear regression model on this dataset. 
Linear regression model will learn that 0 that 1 comes in between 0 and 2. And, the distance between 0 to 1 or 1 to 2 is equal. 
But, this is not the case for qualitative measure, we cannot quantify the blue eye color comes after brown eye colour. 
In case of, binary classification, we can solve the problem with regression but the decision boundary would look like this.




# Intuition behind Logistics Regression

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.datasets import load_digits
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt

import numpy as np
```

```python

mnist = fetch_openml('mnist_784')
```

```python
mnist.data.shape,mnist.target.shape
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(np.array(mnist.data),np.array(mnist.target), test_size=0.33)

x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(x_train[0:5], y_train[0:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (28,28)), )
    plt.title(f'Training: \n{label[0]}')
```

```python
log_reg= LogisticRegression(solver='lbfgs',max_iter=200, verbose=1)
```

```python
log_reg.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
import seaborn as sns
```

```python
y_pred=log_reg.predict(x_test)
```

```python
score=log_reg.score(x_test,y_test)
print(score)
```

```python
preds= log_reg.predict(x_test)
cmtrx= confusion_matrix(y_test,preds)
cmtrx
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cmtrx,annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r')
plt.xlabel('Predicted Label')
plt.ylabel('Actual label')
plt.show()
```
